**TASK**

```python
SHOA= float(input(" Enter (SCORE Prelim) Hands-on Activity: "))
THOA= float(input(" Enter (TOTAL Prelim) Hands-on Activity: "))
RHOA= (SHOA/THOA)
HOA = (RHOA*50)
if HOA > 50:
  print("Invalid Input")
  sys.exit()
print("\n Hands-on Activity Grade: ",HOA)

SQ= float(input("\n Enter (SCORE Prelim) Quiz: "))
TQ= float(input(" Enter (TOTAL Prelim) Qiuz: "))
RQ= (SQ/TQ)
Q = (RQ*30)
if Q > 30:
  print("Invalid Input")
  sys.exit()
print("\n Quiz Grade: ",Q)

SAss= float(input("\n Enter (SCORE Prelim) Ass: "))
TAss= float(input(" Enter (TOTAL Prelim) Ass: "))
RAss= (SAss/TAss)
Ass = (RAss*20)
if Ass > 20:
  print("Invalid Input")
  sys.exit()
print("\n Assignment Grade: ",Ass)

PCS= (HOA+Q+Ass)
print("\n Prelim Class Standing:",PCS)

SPE= float(input("\n Enter (SCORE Prelim) Exam: "))
TPE= float(input(" Enter (TOTAL Prelim) Exam: "))
PE = (SPE/TPE)*(100)
if PE > 100:
  print("Invalid Input")
  sys.exit()
print("\n Prelim Exam Grade: ",PE)

PG= (PCS*.5)+(PE*.5)
print("\nPRELIM GRADE: ",PG)

MG= float(input("\n   Enter (MIDTERM GRADE): "))
if MG > 100:
  print("Invalid Input")
  sys.exit()
print("\nMIDTERM GRADE: ",MG)
RPG= (PG/100)
RMG = (MG/100)
FG = (RPG*.5)+(RMG*.5)

print("\nFINAL GRADE: ",FG*100)
if FG*100 >= 70:
  print("PASSED")
else:
  print("FAILED")
```

**SUPPLIMENTARY**

```python
def get_grade(score, total, weight):
    grade = (score / total) * weight
    if grade > weight:
        print("Invalid Input")
        sys.exit()
    return grade

name = input("Student Name: ")
section = input("Section: ")
print("\n")

HOA = get_grade(float(input(" Enter (SCORE Prelim) Hands-on Activity: ")),
                float(input(" Enter (TOTAL Prelim) Hands-on Activity: ")), 50)
print("\n Hands-on Activity Grade:", HOA)

Q = get_grade(float(input("\n Enter (SCORE Prelim) Quiz: ")),
              float(input(" Enter (TOTAL Prelim) Quiz: ")), 30)
print("\n Quiz Grade:", Q)

Ass = get_grade(float(input("\n Enter (SCORE Prelim) Assignment: ")),
                float(input(" Enter (TOTAL Prelim) Assignment: ")), 20)
print("\n Assignment Grade:", Ass)

PCS = HOA + Q + Ass
print("\n Prelim Class Standing:", PCS)

PE = get_grade(float(input("\n Enter (SCORE Prelim) Exam: ")),
               float(input(" Enter (TOTAL Prelim) Exam: ")), 100)
print("\n Prelim Exam Grade:", PE)

PG = (PCS * 0.5) + (PE * 0.5)
print("\nPRELIM GRADE:", PG)

MG = float(input("\n  Enter (MIDTERM GRADE): "))
if MG > 100:
    print("Invalid Input")
    sys.exit()
print("\nMIDTERM GRADE:", MG)

FG = ((PG / 100) * 0.5) + ((MG / 100) * 0.5)
print("\nFINAL GRADE:", FG * 100)
print("PASSED" if FG * 100 >= 60 else "FAILED")

print("\nSUMMARY:")
print("Student Name:", name)
print("Section:", section)
print("\nPrelim Grade:", PG)
if PG >= 99 <100:
  print("Prelim Grade: 1.00")
elif PG >= 95 <98:
  print("Prelim Grade: 1.25")
elif PG >= 90 <94:
  print("Prelim Grade: 1.50")
elif PG >= 85 <89:
  print("Prelim Grade: 1.75")
elif PG >= 80 <84:
  print("Prelim Grade: 2.00")
elif PG >= 75 <79:
  print("Prelim Grade: 2.25")
elif PG >= 70 <74:
  print("Prelim Grade: 2.50")
elif PG >= 65 <69:
  print("Prelim Grade: 2.75")
elif PG >= 60 <64:
  print("Prelim Grade: 3.00")
else:
  print("Prelim Grade: FAILED")
print("\nMidterm Grade:", MG)
if MG >= 99 <100:
  print("Midterm Grade: 1.00")
elif MG >= 95 <98:
  print("Midterm Grade: 1.25")
elif MG >= 90 <94:
  print("Midterm Grade: 1.50")
elif MG >= 85 <89:
  print("Midterm Grade: 1.75")
elif MG >= 80 <84:
  print("Midterm Grade: 2.00")
elif MG >= 75 <79:
  print("Midterm Grade: 2.25")
elif MG >= 70 <74:
  print("Midterm Grade: 2.50")
elif MG >= 65 <69:
  print("Midterm Grade: 2.75")
elif MG >= 60 <64:
  print("Midterm Grade: 3.00")
else:
  print("Midterm Grade: FAILED")
print("\nFinal Grade:", FG * 100)
if (FG * 100) >= 99 <100:
  print("Final Grade: 1.00")
if (FG * 100) >= 95 <98:
  print("Final Grade: 1.25")
if (FG * 100) >= 90 <94:
  print("Final Grade: 1.50")
if (FG * 100) >= 85 <89:
  print("Final Grade: 1.75")
if (FG * 100) >= 80 <84:
  print("Final Grade: 2.00")
if (FG * 100) >= 75 <79:
  print("Final Grade: 2.25")
if (FG * 100) >= 70 <74:
  print("Final Grade: 2.50")
if (FG * 100) >= 65 <69:
  print("Final Grade: 2.75")
if (FG * 100) >= 60 <64:
  print("Final Grade: 3.00")
else:
  print("Final Grade: FAILED")
```
